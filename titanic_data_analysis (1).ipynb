```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('train.csv')
print(df.head())
```

```python
df.isnull().sum()
```

```python
df['Age'].fillna(df['Age'].median(), inplace=True)
```

```python
df.drop(columns=['Cabin'], inplace=True)
```

```python
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)
```

```python
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['Sex'] = df['Sex'].astype('category')
df['Embarked'] = df['Embarked'].astype('category')
df['Pclass'] = df['Pclass'].astype('category')
```

```python
df.columns
```

```python
# a couple of things we can visualise with the attributes we have

# survived by sex
# survived by age
# survival count like how many survived
# survival by classsurvival by Embarked
```

```python
#let us do the survival count first :p
```

```python
sns.countplot(x='Survived', hue='Survived', data=df, palette='Greens', legend=False)
plt.title('Titanic Survival Count')
plt.xlabel('Survival Status')
plt.ylabel('Number of Passengers')
plt.xticks([0, 1], ['Did Not Survive', 'Survived'])
plt.show()
```

```python
sns.countplot(x='Sex', hue='Survived', data=df, palette='BuPu')
plt.title('Survival by Gender')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
sns.countplot(x='Pclass', hue='Survived', data=df, palette = 'flare')
plt.title('Survival by Class')
plt.legend(title='Survived', labels=['No', 'Yes'])
plt.show()
```

```python
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Feature Correlation Heatmap')
plt.show()
```

```python
sns.violinplot(x='Pclass', y='Age', hue='Survived', data=df, split=True)
plt.title('Survival by Age and Class')
plt.show()
```

### Insights from Titanic Data Analysis

- **Gender:** Women had a much higher survival rate than men.
- **Passenger Class:** First-class passengers survived at a higher rate than second and third class.
- **Age:** Children and younger passengers had better chances of survival.
